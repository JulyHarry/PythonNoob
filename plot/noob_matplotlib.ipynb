```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
from matplotlib import style

# 设置Latex风格
mp.rcParams['text.usetex'] = False
# 设置图像大小
plt.figure(figsize=(5,5), dpi=70)

# 描绘单点
x, y = 1, 100
plt.scatter(x, y, 50)

# 保存图像
# plt.savefig('myplot_singleplot.pdf', dpi=700)
plt.show()
```

```python
# 设置图像大小
plt.figure(figsize=(5,5), dpi=70)
# 设置风格
style.use('ggplot')

# 描绘单点
x, y = 1, 100
plt.scatter(x, y, 50)

# 保存图像
plt.savefig('myplot_singleplot.pdf', dpi=700)
plt.show()
```

```python
# 描绘多点
x_list, y_list = [], []
for i in range(1, 20):
    x_list.append(i)
    y_list.append(np.sin(i))
plt.scatter(x_list, y_list, 50)
plt.savefig('myplot_multiplots.png', dpi=700)
plt.show()
```

```python
# 多点连线
x_list, y_list = [], []
for i in range(1, 20):
    x_list.append(i)
    y_list.append(np.sin(i))
plt.plot(x_list, y_list, 1)
plt.savefig('myplot_multiplotsplot.png', dpi=700)
plt.show()
```

```python
# 多点连线
x_list, y_list = [], []
for i in range(1, 20):
    x_list.append(i)
    y_list.append(np.sin(i))
plt.plot(x_list, y_list, color='grey', marker='o',markersize=8,linewidth=4,markerfacecolor='red',markeredgecolor='green',markeredgewidth=2)
plt.savefig('myplot_multiplotsplot_cus.png', dpi=700)
plt.show()
```

```python
# 绘制函数
x_func = np.linspace(1,10,100)
y_func = np.sin(x_func)
plt.plot(x_func, y_func)
plt.savefig('myplot_function.png', dpi=700)
plt.show()
```
